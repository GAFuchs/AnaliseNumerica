Dado um polinÃ´mio descrito na forma:

$p(x) = a_0 + a_1 x + a_2 x^2 + ... + a_n x^n$

Implemente uma funÃ§Ã£o em python que avalie de forma eficiente esse polinÃ´mio(ou seja, fazendo o menor nÃºmero de multiplicaÃ§Ãµes e somas).

```python
def evalpol (a, x): 
    
    n = len(a)
    if(n > 1):
        p = 0.0
        p = a[0] + x * evalpol(a[1:n],x)
        return p
    else:
        return a[0]

a = [1.0, 1.0, 1.0]
print(evalpol(a, 2.0))
```
